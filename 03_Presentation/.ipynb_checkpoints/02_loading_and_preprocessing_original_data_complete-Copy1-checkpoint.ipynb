```python
# Check with Dicky-Fuller test (DF-test) provides a more accurate way to measure
#   the stationarity of our series. ("Statistic Test" lower than all "Critical Value")
#   CONCLUSION: NOT stationary series
```

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
```

```python
# Perform Dickey-Fuller test:
# Default: AIC, with regression argument of constant drift (deriva) --> ('c'), centering the serie in 0.
print ('Results of Dickey-Fuller Test:')
df_series_tiempo_DickeyFuller = df_series_tiempo.iloc[:,0].values
dftest_DF = adfuller(df_series_tiempo_DickeyFuller, autolag='AIC')
#regression=‘ctt’ : constant, and linear and quadratic trend.
#autolag= ‘t-stat’ based choice of maxlag. Starts with maxlag and drops a lag until the t-statistic on the last lag length is significant using a 5%-sized test.
#store=If True, then a result instance is returned additionally to the adf statistic. Default is False.
#regresults=If True, the full regression results are returned. Default is False.
#dftest_DF = adfuller(df_series_tiempo_DickeyFuller, maxlag=None, regression='ctt', autolag='t-stat', store=True, regresults=True)

dfoutput_DF = pd.Series(dftest_DF[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest_DF[4].items():
    dfoutput_DF['Critical Value (%s)'%key] = value
print(dfoutput_DF)

# The result indicates that the series is NOT stationary, not being able to reject the null hypothesis 
#   at 0.05 (5%) by p-value (without accepting the alternative hypothesis that stationarity exists)
#   Although (aunque) the critical value of the tests (Statistic Test), is smaller than our value level of 
#   significance of .5 (5%), so we do not have enough evidence to conclude that the unit root does not exist.
```
